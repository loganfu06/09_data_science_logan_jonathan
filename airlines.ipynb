```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
df_data = pd.read_excel('airlines.xlsx')
df_data
```

```python
df_data = df_data.dropna()
```

```python
df_data.isnull().values.all()
```

```python
df_data
```

```python
df_data['Date_of_Journey'] = pd.to_datetime(df_data['Date_of_Journey'])
df_data
```

```python
df_data['Dep_Time'] = pd.to_datetime(df_data['Dep_Time']).dt.time
df_data
```

```python
df_data['Arrival_Time'] = pd.to_datetime(df_data['Arrival_Time']).dt.time
df_data
```

```python
df_data['Duration'] = pd.to_timedelta(df_data['Duration'])
df_data
```

```python
df_data['Journey_day'] = df_data['Date_of_Journey'].apply(lambda x: x.day)
df_data['Journey_month'] = df_data['Date_of_Journey'].apply(lambda x: x.month)
df_data['Journey_year'] = df_data['Date_of_Journey'].apply(lambda x: x.year)
df_data
```

```python
df_data['Dep_Time_hour'] = df_data['Dep_Time'].apply(lambda x: x.hour)
df_data['Dep_Time_minute'] = df_data['Dep_Time'].apply(lambda x: x.minute)
df_data['Arrival_Time_hour'] = df_data['Arrival_Time'].apply(lambda x: x.hour)
df_data['Arrival_Time_minute'] = df_data['Arrival_Time'].apply(lambda x: x.minute)
df_data
```

```python
def get_description(time):
    if (time >= datetime.strptime('4:00:00','%H:%M:%S').time()) and (time < datetime.strptime('8:00:00','%H:%M:%S').time()):
        return 'Early Morning'
    elif (time >= datetime.strptime('8:00:00','%H:%M:%S').time()) and (time < datetime.strptime('12:00:00','%H:%M:%S').time()):
        return 'Morning'
    elif (time >= datetime.strptime('12:00:00','%H:%M:%S').time()) and (time < datetime.strptime('16:00:00','%H:%M:%S').time()):
        return 'Noon'
    elif (time >= datetime.strptime('16:00:00','%H:%M:%S').time()) and (time < datetime.strptime('20:00:00','%H:%M:%S').time()):
        return 'Evening'
    elif (time >= datetime.strptime('20:00:00','%H:%M:%S').time()) and (time <= datetime.strptime('23:59:00','%H:%M:%S').time()):
        return 'Night'
    elif (time >= datetime.strptime('00:00:00','%H:%M:%S').time()) and (time < datetime.strptime('4:00:00','%H:%M:%S').time()):
        return 'Late Night'

df_data['dep_description'] = df_data['Dep_Time'].apply(get_description)
df_data
```

```python
df_groupby = df_data.groupby('dep_description').size()
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(df_groupby.index, df_groupby)
plt.ylabel('Number of flights')
plt.xlabel('Departure Time Description')
```

```python
df_data['Duration_hours'] = df_data['Duration'].apply(lambda x: x.seconds // 3600)
df_data['Duration_mins'] = df_data['Duration'].apply(lambda x: (x.seconds // 60) % 60)
df_data['Duration_total_mins'] = df_data['Duration'].apply(lambda x: x.seconds // 60)
df_data
```

```python
plt.scatter(df_data['Duration_total_mins'], df_data['Price'])
plt.xlabel('Total minutes of flight')
plt.ylabel('Price')

m,b = np.polyfit(df_data['Duration_total_mins'], df_data['Price'], 1)
plt.plot(df_data['Duration_total_mins'], m * df_data['Duration_total_mins'] + b, color='#FF0000')
# As the total minutes of the flight increase, the price of the flight increases
```

```python
colors = df_data['Total_Stops'].map({'1 stop':'#FF0000', '2 stops':'#0000FF', '3 stops':'#00FF00', '4 stops':'#FFFF00', 'non-stop':'#000000'})
plt.scatter(df_data['Duration_total_mins'], df_data['Price'], c=colors)
plt.xlabel('Total minutes of flight')
plt.ylabel('Price')
# Most flights are for 1 stop. Non-stop flights are mainly concentrated under 200 minutes.
```

```python
df_jet_airways = df_data.groupby('Airline').get_group("Jet Airways")
final_data = df_jet_airways.groupby('Route').size().sort_values(ascending=False).head(10)
plt.figure(figsize=(28,6))
plt.bar(final_data.index, final_data)
plt.xlabel('Route')
plt.ylabel('Number of flights')
plt.title('Most used routes by Jet Airways')

# CCU -> BOM -> BLR is the most used route by Jet Airways
```
